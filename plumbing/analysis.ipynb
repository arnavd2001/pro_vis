```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
world = [
    ("result_seq_quad.out", "Quadratic", "b", ""),
    ("result_seq_lin.out", "Linear", "g", ""),
    ("result_seq_threads.out", "Quadratic + Threads", "y", ""),
    ("result_seq_lin_threads.out", "Linear + Threads", "cyan", "--"),
    ("result_seq_cuda.out", "Quadratic + CUDA", "orange", "")
]

for fname, label, c, fmt in world:
    df = pd.read_csv(fname)
    group = df.groupby(by="psize")
    plt.errorbar(group.mean().index, group.mean().values, yerr=group.std().values*2, barsabove=True,
                 c=c, label=label, ecolor="black", linewidth=4, fmt=fmt)

plt.ylabel("Execution time (s)")
plt.xlabel("Problem size (# amino acids)")
plt.legend()
plt.savefig("fig_all.pdf")
plt.savefig("fig_all.png", dpi=150)
plt.show()
```
